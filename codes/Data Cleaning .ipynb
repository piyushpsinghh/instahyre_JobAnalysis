```python
import pandas as pd
```

```python
jobs_df=pd.read_csv("E:\instahyre\JOBS2.csv")
jobs_df.dropna(inplace=True)
jobs_df
```

```python
company_df = pd.read_csv("E:\instahyre\COMPANY2.csv")
company_df.dropna(inplace=True) 
company_df
```

```python
details_df=pd.read_csv("E:\instahyre\Details1.csv")
details_df.dropna(inplace=True)
details_df
```

```python
details_df=details_df.assign(Skills=details_df['Skills'].str.split(',')).explode('Skills')
details_df[['Details_ID','Skills']].to_csv('E:\instahyre\Det_Skills.csv')
```

```python
merged_df = jobs_df.merge(company_df, left_on='Company_ID', right_on='Company_ID')
```

```python
locations_df = merged_df.assign(Location=merged_df['Location'].str.split(',')).explode('Location')
locations_df.head(20)
```

```python
result_df = locations_df[['Company_ID', 'Company_Name','Location' ]]
result_df = result_df.sort_values(by='Company_ID')
result_df=result_df.drop_duplicates(keep='first')
result_df.to_csv('E:\instahyre\Comp_Loc.csv')
```
