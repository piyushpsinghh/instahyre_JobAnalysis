```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import time
driver.get("https://www.instahyre.com/search-jobs/")

# all the the link of top 20 jobs 
# each page  product link 
final_list_link = set()

def each_page_links():
    element = driver.find_elements(By.ID,'employer-profile-opportunity')
    for i in element:
        final_list_link.add(i.get_attribute('ng-href'))

for i in range(30):
    each_page_links()
    button = driver.find_element("xpath",'//*[@ng-click="nextPage()"]')
    if button:
        button.click()
        time.sleep(2)

#curr_cum,curr_loc,exp_year,hr_name,skills,founded_yr,size,desg
```

```python
final_tist_link = final_list_link.discard('https://www.instahyre.com/job-244768-backend-developer-at-jktech-bangalore/')
```

```python
final_list_link.discard('https://www.instahyre.com/job-244768-backend-developer-at-jktech-bangalore/')
curr_cum = []
curr_loc =  []
exp_year = []
hr_name = []
skills = []
founded_yr=[]
size=[]
desg = []
for link in final_list_link:
    driver.get(link)
    driver.sleep(5)
    
    try:
        com = driver.find_element(By.XPATH, '//*[@id="floating-header"]/div[1]/h2').t
        curr_cum.append(com)

        x=driver.find_element(By.XPATH, "//div[@class='job-locations']/span").get_attribute("innerHTML").split(">")[-1]
        curr_loc.append(x)

        year_exp=driver.find_element(By.XPATH, "//span[@class='experience ng-binding']").get_attribute("innerHTML")
        exp_year.append(year_exp.split(">")[-1])

        ui=driver.find_element(By.ID,"job-skills-description")
        li=ui.find_elements(By.TAG_NAME,"li")
        Skill_Set=""
        for skill in li:
            if Skill_Set=="":
                Skill_Set=Skill_Set+skill.get_attribute("innerHTML")
            else:
                Skill_Set=Skill_Set+","+skill.get_attribute("innerHTML")
        skills.append(Skill_Set)
        
        fd =driver.find_element(By.XPATH, "//div[@ng-if='employer.company_founded']").get_attribute("innerHTML").split(">")[-1][-4:]
        founded_yr.append(fd[-4:])

        #m = driver.find_element(By.XPATH, "//div[@ng-if='employer.employee_count']").get_attribute("innerHTML")
        m = employees_count=driver.find_element(By.XPATH, "//div[@ng-if='employer.employee_count']").get_attribute("innerHTML")
        size.append(m)        

        hr = driver.find_element(By.XPATH, "//span[@class='rec-name']").get_attribute("innerHTML").split(">")[-1]
        hr_name.append(hr)

        dg = driver.find_element(By.XPATH, "//div[@class='profile-info']/h1").get_attribute("innerHTML").split("(")[0]
        desg.append(dg)
    
    except NoSuchElementException:
        print()
        

result=[]
result.append(curr_cum)
result.append(curr_loc)
result.append(exp_year)
result.append(hr_name)
result.append(skills)
result.append(founded_yr)
result.append(size)
result.append(desg)
```
